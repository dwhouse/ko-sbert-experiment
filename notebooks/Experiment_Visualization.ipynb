```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
categories = ["STS Only", "NLI Only", "STS->NLI", "NLI->STS"]
pearson_scores = [
    0.8161464939414811,
    0.8265220484558593,
    0.8376511958206269,
    0.8492899426478029
]
spearman_scores = [
    0.8142776748758852,
    0.8378181182100715,
    0.8489929165462522,
    0.8554136874109158
]

x = np.arange(len(categories))  # x-axis locations
width = 0.35  # Width of the bars

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width/2, pearson_scores, width, label='Pearson Cosine', color='#1f77b4', edgecolor='black', linewidth=1.2)
rects2 = ax.bar(x + width/2, spearman_scores, width, label='Spearman Cosine', color='#ff7f0e', edgecolor='black', linewidth=1.2)

# Adding text for labels, title, and custom x-axis tick labels
ax.set_xlabel('Models', fontsize=14, fontweight='bold')
ax.set_ylabel('Scores', fontsize=14, fontweight='bold')
ax.set_title('Comparison of Test Scores by Model and Metric', fontsize=16, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=12)
ax.legend(fontsize=12)
ax.set_ylim(0.8, 0.86)  # Adjust y-axis for better visibility of differences

# Annotate bars with their values
def annotate_bars(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 5),  # Offset text slightly above the bar
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10, color='black')

annotate_bars(rects1)
annotate_bars(rects2)

# Add gridlines for better readability
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

# Show the plot
plt.tight_layout()
plt.show()
```
